```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)

ax.plot(x_axis, y_axis, color= "green", marker="D", linewidth=2, label="Boston")
ax.set_ylim(0,45)
ax.set_xlabel("Months")
ax.set_ylabel("Fare ($)")
ax.set_title("PyBer Fare by Month")
ax.grid()

ax.legend()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(y_axis, x_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()

plt.plot(color = "magenta", label="Boston")
plt.legend("Boston")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax = plt.gca()
ax.invert_yaxis()
ax.barh(x_axis, y_axis)

ax.plot(color = "cyan", label = "Chicago")
ax.legend("Boston")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.scatter(y_axis, x_axis, c="red", label= "Chicago")
plt.legend()
plt.xlabel("Months")
plt.ylabel("Fares($)")
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s= [i * 3 for i in y_axis], c="lightseagreen", edgecolor= "black", alpha = .2, linewidth=2, label="Boston")
ax.legend("Boston")
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s= [i * 3 for i in y_axis], c="lightseagreen", edgecolor= "black", alpha = .2, linewidth=2, label="Boston")
ax.legend("Boston")
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%', shadow=True)
plt.figure(figsize=(18, 18))
```
