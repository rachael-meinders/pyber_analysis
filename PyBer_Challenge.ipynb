# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head(10)
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type

total_rides_by_ctype = pyber_data_df.groupby(["type"]).count()["ride_id"]
print(total_rides_by_ctype)
```

```python
# 2. Get the total drivers for each city type
total_drivers_by_ctype = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_by_ctype
```

```python
#  3. Get the total amount of fares for each city type
total_fares_by_ctype = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_by_ctype                                          
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_by_ctype =  pyber_data_df.groupby(["type"]).sum()["fare"] / pyber_data_df.groupby(["type"]).count()["ride_id"]
avg_fare_by_ctype
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver_by_ctype = pyber_data_df.groupby(["type"]).sum()["fare"] / city_data_df.groupby(["type"]).sum()["driver_count"]
avg_fare_per_driver_by_ctype
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame(
    {'Total Rides': total_rides_by_ctype,
      'Total Drivers': total_drivers_by_ctype,
      'Total Fares': total_fares_by_ctype,
      'Average Fare per Ride': avg_fare_by_ctype,
      'Average Fare per Driver': avg_fare_per_driver_by_ctype})

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# A. Read the merged DataFrame
pyber_data_df
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
typedate_df = pyber_data_df.groupby(["date", "type"]).sum()["fare"]
typedate_df
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
typedate_df = typedate_df.reset_index()
typedate_df
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
new_typedate_df = typedate_df.pivot("date","type", "fare")
new_typedate_df
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
new_typedate_df.index = pd.to_datetime(new_typedate_df.index)
new_typedate_df
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
start = pd.Timestamp(2019, 1, 1, 0, 8, 16)
stop = pd.Timestamp (2019, 4, 29, 11, 10, 18)
stop

jan_to_apr28_df = new_typedate_df.loc["2019-01-01":"2019-04-28", :]
jan_to_apr28_df
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
jan_to_apr28_df.index = pd.to_datetime(jan_to_apr28_df.index)
jan_to_apr28_df
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
jan_to_apr28_df.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
by_week_df = jan_to_apr28_df.resample("w").sum()
by_week_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
fig, ax = plt.subplots(figsize = (20,10))
ax.plot(by_week_df.index,
        by_week_df["Urban"],
       label = "Urban")
ax.plot(by_week_df.index,
        by_week_df["Suburban"],
       label = "Suburban")
ax.plot(by_week_df.index,
        by_week_df["Rural"],
       label = "Rural")


ax.legend(fontsize="12", mode="Expanded",
          loc="best", title="City Types")

ax.set_ylabel("Fare ($USD)")
ax.set_xlabel("Date (2019)")
ax.set_title("Total PyBer Fare by City Type")

ax.grid(True)

# x axis formatting
# taken from https://stackoverflow.com/questions/45704366/how-to-change-the-datetime-tick-label-frequency-for-matplotlib-plots
import matplotlib.dates as mdates
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))


# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
# save figure
plt.savefig("analysis/Fig8_challenge.png")
# show
plt.show()
```
